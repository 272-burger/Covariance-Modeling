# Step 1. Get the observations together

* cf. By default, Numpy uses float64 while TensorFlow uses float32.
    * As Numpy does data casting automatically, while TF needs explicit data casting, it is much easier to convert your Numpy data into float32. 

## 1-1. Choose parameter values.

```python
# function that makes covariance matrix
def Sigma(param,k):
    I = np.eye(k, dtype=np.float32)
    O = np.ones([k,k], dtype=np.float32)
    return I*param[0]+(O-I)*param[1]
```

```python
k = 1000 # dimensionality
paramTrue = [1,.1] 
```

```python
# resulting covariance matrix
SigmaTrue = Sigma(paramTrue, k)
print(SigmaTrue)

print('eigenvalues: ', np.linalg.eigvals(SigmaTrue))
```

## 1-2. Generate some synthetic observations. 
We'll generate an array my_data with shape (100, 2). 
* TF vector shape convention: shape(n, p)
    * n = sample indices
    * p = dimensionality

```python
# set the seed so the results are reproducible.
np.random.seed(123)

# Now generate some observations of our random variable.
SigmaHalf = np.linalg.cholesky(SigmaTrue)
U = np.random.normal(0, 1, k).reshape(k, 1).astype(np.float32)

my_data = np.dot(SigmaHalf, U)

print(my_data.shape)
```

```python
# Do a scatter plot of the observations to make sure they look like what we
# expect (higher variance on the x-axis, y values strongly correlated with x)
plt.scatter(x=U, y=my_data, alpha=0.75)
plt.show()
```

* Split the likelihood function into 2 pieces:
    * Data likelihood function: $P(\text{parameters}|\text{data})$
    * Prior likelihood function: $P(\text{parameters})$
        * (I don't want to study what this is right now but) Using Wishart distribution for prior. 
        * The Wishart distribution is the conjugate prior for **estimating the precision matrix** of a multivariate normal with known mean $\mu$. Suppose the prior Wishart parameters are $\nu, V$ and that we have $n$ observations of our multivariate normal, $x_1, \ldots, x_n$. The posterior parameters are $n + \nu, \left(V^{-1} + \sum_{i=1}^n (x_i-\mu)(x_i-\mu)^T \right)^{-1}$
* Note: We're going to be working on precision matrix ($\Sigma^{-1/2}$), so we'll parameterize accordingly.
